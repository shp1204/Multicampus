```python
# python의 기본 파일처리

# 먼저 file handle을 획득해야한다.(= file open)
# 사용할 데이터 파일을 오픈하자
# 읽기용도, 쓰기용도

# open("파일이름","용도")
# 파일은 쥬피터노트북과 같은 위치에 있다.
# file1은 오픈(r), file2는 file1을 복사해서 변경 및 수정(w)
file1=open("student_score.txt","r")
file2=open("backup.txt","w")

# 학생 성적 파일에서 한줄씩 읽어와서
# output console에 출력하고 해당 내용을 backup 파일에 저장

# 파일은 몇줄까지 있는지 모르기 때문에 반복을 몇 번 알지 정확히 알 수 없다.
# 따라서 for문이 아닌 while문으로 조건(파일이 끝날때까지)을 만들어주는 것이 좋다.

while True:
    line = file1.readline()
    print(line)
    if not line:
        break
    
# line이 아무것도 없을 때, 반복문을 탈출한다.

# 결과 : 
# 홍길동,18,7,19

# 최길동,58,70,15

# 아이유,92,94,99

# 김연아,17,20,22

# 이선희,76,25,14

# 강감찬,90,19,71

# 신사임당,31,23,16

# 왜 한줄씩 띄어져서 나올까????
# 눈에 보이지않지만 개행문자(\n)가 있기 때문이다.
# 개행문자는? 한줄씩 띄라는 의미
# print 함수는 출력한 다음 한줄을 띈다
# 홍길동,18,7,10(\n) 

while True:
    line = file1.readline()
    print(line, end="")
    if not line:
        break

# 결과 :
# 홍길동,18,7,19
# 최길동,58,70,15
# 아이유,92,94,99
# 김연아,17,20,22
# 이선희,76,25,14
# 강감찬,90,19,71
# 신사임당,31,23,16

while True:
    line = file1.readline()
    print(line, end="")
    if not line:
        break
    file2.write(line)

file1.close()
file2.close()

# 파일을 닫아줘야 실제로 저장된다!
# 다하고 파일 안닫으면 저장 안됨 ㅠㅠ
```
