```python
# Bool type : bool은 True와 False가 있다.
# and와 or와 not은 논리연산자
# &은 비교연산자, bitwise연산
# 파이썬에서 숫자 0은 False로 간주, 0이 아닌 모든 숫자는 True
# 파이썬에서 False는 숫자 0으로 변환이 가능하다 !
print(bool(0))
# 결과 : False
print(bool(10))
# 결과 : True
print(int(False))
#결과 : 0
#10진수 5를 2진수 => 0101
#10진수 1을 2진수 => 0001
# True&False => 1&0(숫자로변환) => 0001&0000(bit로변환) => 0000(bit 연산결과) => 0(숫자로변환) => False

# 파이썬은 다음의 경우를 False로 간주한다
# 숫자 0, 빈 문자열(""), 빈 리스트([]), 빈 튜플 (())
# 비어있는 dict ( {} ), None
```

```python
# Set type : {}
# dict는 key, value로 구성한다 set는 요소만 들어간다
a = {1, 2, 3}
print(type(a))
# 결과 : <class 'set'>
# set은 순서가 없고 중복을 허용하지 않는 저장장소
a=set([1,2,3,4])
print(a)
# 결과 : {1, 2, 3, 4}
a=set("Hello")
print(a)
# 결과 : {'H', 'o', 'e', 'l'}

#set의 연산
a={1,2,3,4}    # set의 literal
b={3,4,5,6}

# set => 집합 ( 집합연산, 교집합, 합집합, 차집합 )
print(a & b)
# 결과 : {3, 4}
print(a.intersection(b))
# 결과 : {3, 4}
print(a|b)
# 결과 : {1, 2, 3, 4, 5, 6}
print(a.union(b))
# 결과 : {1, 2, 3, 4, 5, 6}
print(a-b)
# 결과 : {1, 2}
print(a.difference(b))
# 결과 : {1, 2}

# add는 set에 원하는 요소 1개 추가 => 겹치는건 추가 안해준다
mySet = {1,2,3,4,5}
mySet.add(7)
print(mySet) 
# 결과 : {1, 2, 3, 4, 5, 7}

# update는 한번에 여러개 요소 추가
mySet = {1,2,3,4,5}
mySet.update({10,100,1000})
print(mySet)
# 결과 : {1, 2, 3, 4, 5, 100, 1000, 10}
mySet.remove(1)
print(mySet)
# 결과 : {2, 3, 4, 5, 100, 1000, 10}
```
