```python
### 문제 15.
# 6자리 이상 9자리 미만의 수를 하나 이용하자
# ex) 234567  6522345
# 수의 중앙을 기준으로 두 개의 수로 분리
# 234567 => (234, 567)
# 6522345 => (652, 345)
# 두개의 수로 분리한 다음 큰 수를 선택하고
# 이 작업을 더 이상 나눌 수 없을 때까지 진행한다.
# ex ) 6522345 => 652 와 345 => 652 => 6 과 2 => 6
# 최종적으로 남은 수는 얼마인가요?

origin = 200195
while origin not in range(0,10):
    str_origin = str(origin)
    len_devide = int(len(str_origin)/2)
    first = int(str_origin[:len_devide])
    second = int(str_origin[-len_devide:])
    if first > second:
        origin = first
    else:
        origin = second    

print(origin)
```
