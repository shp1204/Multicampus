```python
# 객체 지향 프로그래밍(Object Oriented Programming)
# 함수 기반의 프로그래밍(절차적 프로그래밍)을 하면 좋다
# 1970년부터 시작해서 지금까지 이 방식으로 프로그래밍을 하고 있다.
# 장점 : 프로그래밍이 쉬워서 빨리 만들 수 있다. 비용이 적게 든다.
# 단점 : 처음에는 단점이 별로 없어보였는데 1990년대부터 세상이 변하기 시작했다 => 인터넷
# 정보량이 많아지기 시작하면서 세상이 급변하기 시작했다
# 프로그램의 유지보수 요구가 많아지기 시작했다
# 절차적 프로그램으로 프로그램을 작성했더니 유지보수비용 및 절차 문제가 발생하기 시작한다
# 어떤 방식으로 프로그램을 만들면 유지보수 측면에서 이점이 있을까?
# 객체지향 paradigm이 시작된다 = 어떤 프로그래밍 방식인지 인지하기 시작한다
# 현실세계에 잘 부합하는(변화에 잘 적응하는) 프로그램을 만들기 위해 
# 해결해야 하는 문제를 프로그램으로 똑같이 모델링 하면 되지 않을까?

# 은행프로그램을 만들어보자
# 은행프로그램을 구성하는 구성요소를 파악해야한다
# 텔러, 고객, 은행계좌, 지점, 테이블, ,,,
# 분석하는 사람마다 은행을 구성하는 구성요소가 다르다
# 어떻게 상호작용을 하는지 프로그래밍한 것이 객체지향이다.
# 구성요소간의 행위들을 정의하는 프로그래밍 방식, 은행계좌(객체)를 예로 프로그램으로 표현(객체 모델링)
# 객체들은 크게 두 가지로 파악할 수 있다.
# 상태와 행위로 파악할 수 있다.

# 상태는 값으로 표현되는 것, 행위는 동작으로 표현되는 것
# 상태 : 잔액, 예금주, 계좌번호, ,,,,
# 행위 : 출금, 입금, 조회, ,,,, => abstraction(추상화)

# class : 객체 모델링의 수단, 추상 데이터타입
# class 이름은 대문자로 시작한다(관용적)
# 
class Student:
    # class에 정의된 내용(상태,함수)을 보고 constructor가 적합한 data type을 선별하여 정보를 저장해줌
    
    # 상태(값) : 변수로 표현 => 이름, 학과, 학년, 학번, 성별, 나이, ,,, 
    # 상태는 변수로 표현 => properties
    # properties
    s_nation = "" #class variable
   




    # @@@@생성자(constructor)@@@@
  
    def __init__(self,n):
        self.s_name = n   #class variable( 모든 객체가 공유하는 변수 )
        self.s_num = ""   #class variable          
        self.s_major = []
        
        
        
        
        
        
    # 행위(동작) : 함수
    # @@@@method@@@@
    def display(self):
        print("학생이름은 : {}".format(self.s_name))

# => 내부에 두개의 변수를 저장할 수 있는 데이터 타입.

t = Student("홍길동")   
            #=>  이 코드에 의해 만들어진 메모리 저장 공간을 "instance"라고 한다
            #=> 이를 통칭해서 객체라고 부른다.
            #=> class 내부에 정의된 생성자라고 불리는 함수가 호출된다.(constructor)
            #=> 고정된 이름이다 @  def __init__(self.n):

#             self.n 이면 함수형태로 호출할 때 Student("홍길동")이라고 넣어줘야한다.
#             그래야 객체가 만들어지기 때문에 
#             self 면 안넣어도됨
print(t.s_name)

# classdp 변수영역과 함수영역으로 나누고, 변수영역은 class variable(모든객체가 공유하는 변수), 생성자는 class의 객체를 만든다
# method를 정의를 했는데 만약에 인자가 아무것도 안들어가면 ..?

# class로 공간을 만들었다. 객체를 만들면 생성자가 자동으로 호출되면 n값이 instance variable의 객체가 독립적으로 가지는 저장공간에 들어간다

# k = Student("김연아") 라고하면 객체가 한명 더 만들어진다.




students = list()
Student.s_nation = "한국" # 모든 객체가 공유하는 변수
t = Student("홍길동")
print(t.s_name)
print(t.s_nation)      # 한국
t.display()

k = Student("김연아")
print(k.s_name)
print(k.s_nation)      # 한국
k.display()

# class student를 이용해서 만들었다.

# print(Student.s_nation)
```
